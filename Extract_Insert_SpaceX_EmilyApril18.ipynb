```python
# Dependencies
import pandas as pd
import requests
import json
from pprint import pprint
```

```python
# Performing a GET Request and saving the 
# API's response within a variable
url = "https://api.spacexdata.com/v3/rockets"
response = requests.get(url)

all_rockets = response.json()
#pprint(all_rockets)
```

```python
rocket_df = pd.io.json.json_normalize(all_rockets)
rocket_df.head()
```
