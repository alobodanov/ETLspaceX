```python
# Dependencies
import pandas as pd
import requests
import json
from pprint import pprint
```

```python
pd.set_option('display.max_columns', 100)
```

```python
# Performing a GET Request and saving the 
# API's response within a variable
url = "https://api.spacexdata.com/v3/rockets"
response = requests.get(url)

all_rockets = response.json()
#pprint(all_rockets)
```

```python
rocket_df = pd.io.json.json_normalize(all_rockets)
```

```python
rocket_df.head()
```

```python
# rocket_df.count()
# rocket_df.dtypes
```

```python
rocket_df.columns
```

```python
rocket_df_final = rocket_df[['id','rocket_name','rocket_id','cost_per_launch','country', 'description'
                            ,'rocket_type', 'active']].copy()
rocket_df_final.head()
```

```python
# Performing a GET Request and saving the 
# API's response within a variable
url = "https://api.spacexdata.com/v3/launches"
response = requests.get(url)

all_launches = response.json()
# pprint(all_launches)
```

```python
launches_df = pd.io.json.json_normalize(all_launches)
```

```python
launches_df.head()
```

```python
# launches_df.count()
# launches_df.dtypes
```

```python
launches_df.columns
```

```python
launcher_df_final = launches_df[['rocket.rocket_name','flight_number','rocket.rocket_id','launch_date_utc'  
                                ,'details','flight_number','launch_year']].copy()
```

```python
launcher_df_final.head()
```

```python
#Rename the Dataframe columns to match the table columns

launcher_df_final.rename(columns={'rocket.rocket_name':'rocket_name'
                                  ,'flight_number':'launch_pads_id'
                                  ,'rocket.rocket_id':'rocket_id'
                                  ,'launch_date_utc':'date_utc'
                                  ,'launch_year':'year'}, 
                                  inplace=True)
```

```python
# Performing a GET Request and saving the 
# API's response within a variable
url = "https://api.spacexdata.com/v3/launchpads"
response = requests.get(url)

launch_pads= response.json()
#pprint(launch_pads)
```

```python
launch_pads_df = pd.io.json.json_normalize(launch_pads)
```

```python
launch_pads_df.head()
```

```python
launch_pads_df.columns
```

```python
launch_pads_final_df = launch_pads_df[['id', 'location.name', 'location.region', 'status', 'details'
                                      , 'successful_launches']].copy()
```

```python
launch_pads_final_df.head()
```

```python
#Rename the Dataframe columns to match the table columns

launch_pads_final_df.rename(columns={'id':'launch_pads_id'
                                  ,'location.name':'location_name'
                                  ,'location.region':'region'}, 
                                  inplace=True)
```
